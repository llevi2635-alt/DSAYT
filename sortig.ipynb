#### SELECTION SORTING :
By one num comparing all elements and converting list into as to des

```python
##(1Q) A = (5,6,7,1,4,2,3,8,9) convert this list into as to dea
def sorting (nums):
    n = len(nums)
    k = 3
    for i in range (0,n):
        mini = i 
        for j in range (i+1,n):
            if nums[j] < nums[mini] :
                mini = j 
        nums[i],nums[mini] =nums[mini],nums[i]
    return nums [k-1]        
sorting([5,6,7,1,4,2,3,8,9])
```

```python
### Apply Selection Sort only for the first half of the array.
### Input: [10, 3, 8, 5, 2, 7]
### Output (first half sorted): [2, 3, 5, 10, 8, 7]
def selection_sorting (nums):
    n = len(nums)
    half = n //2
    for i in range (half):
        mini = i 
        for j in range (i+1,half):
            if nums [j]  < nums [mini]:
                mini = j 
        nums [i],nums[mini] = nums[mini],nums[i]
    return nums
nums =  [2, 3, 5, 10, 8, 7]
selection_sorting (nums)
```

```python
#Use Selection Sort to sort based on absolute values.
#Input: [-3, 1, -7, 4, -2]
#Output: [1, -2, -3, 4, -7] ##(abs = absolute )
def selection_sort_abs(arr):
    n = len(arr)

    for i in range(n):
        min_index = i
        for j in range(i + 1, n):
            if abs(arr[j]) < abs(arr[min_index]):
                min_index = j

        # swap
        arr[i], arr[min_index] = arr[min_index], arr[i]

    return arr


arr = [-3, 1, -7, 4, -2]
print(selection_sort_abs(arr))
```

```python
##Find the k-th smallest element using Selection Sort logic.
##Input: [9, 4, 6, 2, 8], k = 3
##Output: 6
def selection_sorting(nums,k):
    n = len (nums)
    for i in range (0,k):
        mini = i
        for j in range (i+1,n):
            if nums [j] < nums [mini]:
                mini = j
        nums [i],nums[mini] = nums[mini],nums[i]
    return nums[k-1]
selection_sorting([9, 4, 6, 2, 8],3)
```

```python
## (1Q)Write a program to sort a list using Bubble Sort nums = [5,1,6,8,2,4,9]
nums =[5,1,6,8,2,4,9]
n = len (nums)
for i in range(n-2,-1,-1):
    for j in range (0,i+1):
        if nums[j] > nums [j+1]:
            nums[j],nums[j+1] = nums[j+1],nums[j]
print (nums)
```
