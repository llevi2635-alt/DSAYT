```python
### armstrong number 
def armstrong_number (num):
    number = num
    num = str (num)
    length =len(num)
    total = 0
    for i in num :
        total += int (i)** length
    return number == total
armstrong_number(153)
```

```python
##(1Q) given num n = (5,3,2,2,1,5,5,7,5,10) m = (10,111,1,9,5,6,7,2) numbers which r in m . find that r present in n .
#if they r present print how many times r present 
n = (5,3,2,2,1,5,5,7,5,10)
m = (10,111,1,9,5,6,7,2)
hash_list = [0]*11
for num in n :
    hash_list[num]+=1
for nums in m :
    if nums <0 or nums>10:
        print ('out of range')
    else :
        print (hash_list[nums])
```

```python
##(1Q) given num n = (5,3,2,2,1,5,5,7,5,10) m = (10,111,1,9,5,6,7,2) numbers which r in m . find that r present in n .
#if they r present print how many times r present 
n = (5,3,2,2,1,5,5,7,5,10)
m = (10,111,1,9,5,6,7,2)
dir = {}
for num in n:
    if num in dir :
        dir [num]+=1
    else : 
        dir[num]=1
for num in m :
    if num in dir :
        print(num,"->",dir[num])
    else :
        print (num,"->",0)
```
