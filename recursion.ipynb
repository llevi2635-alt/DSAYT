#### RECURSION FUNCTION :- CALLING ITS OWN FUNCTION IT SELF ITS CALLED RECURSION

```python
##(1Q) print 1 to 5 nums ..?
def one_five (num):
    if num == 6 :
        return 
    print (num)
    one_five (num+1)
one_five(1)
### this is an a head recursion
```

```python
##(2Q) print  A 4 times 
def AAAA (count): 
    if count == 4 :
        return 
    print ("A")
    AAAA(count + 1)
AAAA(0)
```

```python
##Sum of First N Natural Numbers 
def sum_n(n):
    if n == 1:          # Base case
        return 1
    return n + sum_n(n - 1)   # Recursive call

# Example
print(sum_n(5))   # Output: 15
```

```python
##Power of a Number (Example: 2^5 = 32)
def power(x, n):
    if n == 0:          # Base case
        return 1
    return x * power(x, n - 1)

# Example
print(power(2, 5))   # Output: 32


"""##power(2,5)
= 2 * power(2,4)
= 2 * (2 * power(2,3))
= 2 * (2 * (2 * power(2,2)))
= 2 * (2 * (2 * (2 * power(2,1))))
= 2 * (2 * (2 * (2 * (2 * power(2,0)))))
= 2 * 2 * 2 * 2 * 2 * 1"""
### = 32
```

```python
###27. Remove Element (LC)
def removeElement(nums, val):
    k = 0  # pointer for valid elements
    
    for i in range(len(nums)):
        if nums[i] != val:
            nums[k] = nums[i]
            k += 1
    
    return k
nums = [3,2,2,3]
val = 3
removeElement(nums,val)
```
